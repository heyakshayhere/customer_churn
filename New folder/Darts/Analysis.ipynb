```python
import pandas as pd 
import numpy as np
import darts 
import yfinance as yf 
import pandas_ta as ta
%matplotlib inline
%autosave 1
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

Importing data

```python
ticker = yf.download(tickers='^NSEI',interval='1D')
ticker
```

```python
ticker.drop(['Adj Close','Volume'],axis = 1,inplace= True)
ticker
```

```python
adx_df = ta.adx(ticker['High'],ticker['Low'],ticker['Close'])
adx_df=adx_df.drop(['ADX_14'],axis=1)
adx_df
```

```python
data = pd.DataFrame(ticker['Close']).dropna()
data = data.reset_index(drop=True)
data
```

```python
data.plot();
```

```python
train,val= data[:500],data[500:]
```

```python
train
```

```python
train.plot()
val.plot();
```

Converting into timeseries

```python
from darts.timeseries import TimeSeries
```

```python
train = TimeSeries.from_dataframe(train, time_col=None, value_cols=None, fill_missing_dates=False, freq='B', fillna_value=False)
train
```

```python
from darts.dataprocessing.transformers import Scaler
```

```python
train_scaling = Scaler()
train_scaled = train_scaling.fit_transform(train)
train_scaled
```

```python
from darts.models.forecasting.nbeats import NBEATSModel
```

```python
help(NBEATSModel)
```

```python
model = NBEATSModel(
    input_chunk_length=24,
    output_chunk_length=12,
    random_state=0,
    n_epochs=100,
    batch_size = 500,
    generic_architecture=True,
    num_stacks=9,
    num_blocks=9,
    num_layers=9, #for 2 blocks it had 8layers
    trend_polynomial_degree=3 
)
```

```python
model.fit(train_scaled,verbose=True)
```

```python
n=100
forecast = model.predict(series=train_scaled,n=n,n_jobs=-1)
forecast.plot();
val[:n].plot();
plt.rcParams["figure.figsize"] = (25,5)
```

```python
output = Scaler.inverse_transform(forecast)
```
