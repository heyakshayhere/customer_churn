```python
import time
start = time.time()
```

```python
import pandas as pd 
import numpy as np
import darts 
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from darts.timeseries import TimeSeries
from darts.models import Theta
from darts.metrics import mape
import seaborn as sns
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,10)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#Range for theta
tl=-50
tr=50
r=1000
#forecast horizons
f1=50
f2=5
#Backtesting
b=10
```

```python
# #Range for theta
# tl=-50
# tr=50
# r=10
# #forecast horizons
# f1=50
# f2=5
# #Backtesting
# b=2
```



<hr/>



```python
list_of_names = ['^NSEI','^NSEBANK','^CNXIT']
  
```

```python
%%time

for name in list_of_names:
  print('   ')

  print(f"Predicting for {name}")

  print('   ')

  Time_Frame = '1d'

  print(f"TimeFrame Selected = {Time_Frame}")

  print('   ')

  #downloading dataset from yfinance

  print('Downloading the data...')

  print('   ')

  ticker = yf.download(tickers=name,interval=Time_Frame)

  print('   ')

  print('Downloading Done!')

  print('  ')

  print(f"Data has {ticker.shape[0]} length")


  print('   ')

  print('Cleaning the raw data...')

  print('   ')

  ticker.drop(['Adj Close','Volume'],axis = 1,inplace= True)
  data = pd.DataFrame(ticker['Close']).dropna()
  data = data.reset_index(drop=True)

  print('Fitting into valid timeseries format...')

  print('   ')
  #converting into timeseris format

  data_timeseries = TimeSeries.from_dataframe(data, time_col=None, value_cols=None, fill_missing_dates=False, freq='B', fillna_value=False)

  #Splitting into training and testing
  length = int(len(data)*(.80))
  train, val = data_timeseries[:length], data_timeseries[length:]

  # Search for the best theta parameter, by trying different values

  print('Searching the best theta...')

  print('   ')

  thetas = np.linspace(tl, tr, r)

  best_mape = float('inf')
  best_theta = 0

  for theta in thetas:
      model = Theta(theta)
      model.fit(train)
      pred_theta = model.predict(len(val))
      res = mape(val, pred_theta)

      if res < best_mape:
          best_mape = res
          best_theta = theta

  best_theta_model = Theta(best_theta)
  best_theta_model.fit(train)
  pred_best_theta = best_theta_model.predict(len(val))

  print('The MAPE for the model is: {:.4f} %, with theta = {}.'.format(mape(val, pred_best_theta), best_theta))


  #backtesting 
  print('   ')

  print('Backtesting...')

  print('   ')

  #must between 99 and 1

  test_for_recent_percentage_data = b  

  forecast_horizon_1 = f1

  print(f"Forecast horizon selected = {forecast_horizon_1} upcoming predictions ")

  historical_fcast_theta = best_theta_model.historical_forecasts(data_timeseries,
                                                                start=(100-test_for_recent_percentage_data)/100,
                                                                forecast_horizon=forecast_horizon_1,
                                                                verbose=True)

  print('MAPE for backtesting = {:.4f}%'.format(mape(historical_fcast_theta, data_timeseries)))

  print(' ')

  #my model 
  best_theta_model = Theta(best_theta)
  best_theta_model.fit(data_timeseries)
  prediction_1 = best_theta_model.predict(forecast_horizon_1)

  #backtesting 
  print('   ')

  print('Backtesting...')

  print('   ')

  #must between 99 and 1

  test_for_recent_percentage_data = b  

  forecast_horizon_2 = f2

  print(f"Forecast horizon selected = {forecast_horizon_2} upcoming predictions ")

  historical_fcast_theta = best_theta_model.historical_forecasts(data_timeseries,
                                                                start=(100-test_for_recent_percentage_data)/100,
                                                                forecast_horizon=forecast_horizon_2,
                                                                verbose=True)

  print('MAPE for backtesting = {:.4f}%'.format(mape(historical_fcast_theta, data_timeseries)))

  print(' ')

  #my model 
  best_theta_model = Theta(best_theta)
  best_theta_model.fit(data_timeseries)
  prediction_2 = best_theta_model.predict(forecast_horizon_2)

  print(' ')
  # plotting the prediction

  prediction_1.plot(new_plot=True,label=f"prediction for next {forecast_horizon_1 * int(Time_Frame[:-1])} {Time_Frame[-1:]}",color='orange')
  prediction_2.plot(label=f"prediction for next {forecast_horizon_2 * int(Time_Frame[:-1])} {Time_Frame[-1:]}",color = 'black')
  data_timeseries[-300:].plot(label=f"Actual Data {Time_Frame}",color = 'blue')
  plt.title(name)
  plt.show();

  print('   ')
```

```python
end = time.time()
intervals = (
    ('weeks', 604800),  # 60 * 60 * 24 * 7
    ('days', 86400),    # 60 * 60 * 24
    ('hours', 3600),    # 60 * 60
    ('minutes', 60),
    ('seconds', 1),
)

def display_time(seconds, granularity=2):
    result = []

    for name, count in intervals:
        value = seconds // count
        if value:
            seconds -= value * count
            if value == 1:
                name = name.rstrip('s')
            result.append("{} {}".format(value, name))
    print(f"The time taken to run the notebook is :")
    print(' '.join(result[:granularity]))

display_time(end-start)
```
